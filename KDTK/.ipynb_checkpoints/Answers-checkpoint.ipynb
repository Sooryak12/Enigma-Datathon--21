```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegressionCV,LinearRegression
```

```python
df=pd.read_csv("Fin. Stats KTDK-2 - Fin. Stats.csv")
```

```python
df
```

```python
df.describe()
```

```python
y=df["Bankrupt?"]
```

```python
X=df.drop(["ID","Bankrupt?"],axis=1)
```

```python
lr=LogisticRegressionCV(cv=5,max_iter=1000,fit_intercept=False,random_state=42,penalty="l1",solver="saga")
lr.fit(X,y)
```

```python
lr.coef_
```

```python
df.drop("ID",axis=1).corr()["Bankrupt?"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X,y)
pred=dt.predict(X)
print(f1_score(y,pred))
```

```python
rf=RandomForestClassifier()
rf.fit(X,y)
```

```python
pred=rf.predict(X)
print(f1_score(y,pred))
```

```python
dt.feature_importances_
```

```python
weights=np.array(rf.feature_importances_)
```

```python
val=np.array(df.drop(["ID","Bankrupt?"],axis=1).iloc[1,:].values)
```
